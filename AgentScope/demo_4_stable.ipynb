```python
import warnings; warnings.filterwarnings("ignore")
import os
import openai
import agentscope
import httpx

from dotenv import load_dotenv
from openai import OpenAI
from agentscope.model import (OllamaChatModel, OpenAIChatModel, DashScopeChatModel)

print("OpenAI version:", openai.__version__)
print("AgentScope version:", agentscope.__version__)
```

```python
curr_path = os.getcwd()
parent_path = os.path.dirname(curr_path)
dotenv_path = os.path.join(parent_path, ".env")
load_dotenv(dotenv_path=dotenv_path)
API_KEY = os.getenv("ARK_API_KEY")
```

# 基础运用

```python
model = OpenAIChatModel(
    model_name="doubao-seed-2-0-pro-260215",
    api_key=API_KEY,
    client_kwargs={
        # "http_client": httpx.AsyncClient(verify=False),
        "base_url": "https://ark.cn-beijing.volces.com/api/v3",
    },
    generate_kwargs={
        "temperature": 0.5,
        "top_p": 0.5,
        # "extra_body": {"stream_options": {"exclude": ["reasoning"]}}
    },
    stream=False,
)
```

```python
messages = [
    {"role": "user", "content": "请介绍一下自己。"},
]

res = await model(messages=messages)
print(res)
```
