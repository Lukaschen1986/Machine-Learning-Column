# 基础运用

```python
import os
import agentscope

from agentscope.model import OllamaChatModel

print("AgentScope version:", agentscope.__version__)
```

```python
model = OllamaChatModel(
    model_name="qwen3:8b",
    host="http://localhost:11434",
    # enable_thinking=False,
    stream=False,
)
```

```python
res = await model(
    messages=[
        {"role": "user", "content": "请介绍一下自己。"},
    ]
)

print(res)
```
